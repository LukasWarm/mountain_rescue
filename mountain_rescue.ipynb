## Exploring stop words with nltk 

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stop= set(stopwords.words('english'))
```

```python
data['DESCRIPTION'] = data['DESCRIPTION'].astype(str)
```

```python
corpus=[]
reports= data['DESCRIPTION'].str.split()
reports=reports.values.tolist()
corpus=[word for i in reports for word in i]
```

```python
def plot_top_stopwords_barchart(text):
    stop=set(stopwords.words('english'))
    
    new= text.str.split()
    new=new.values.tolist()
    corpus=[word for i in new for word in i]
    from collections import defaultdict
    dic=defaultdict(int)
    for word in corpus:
        if word in stop:
            dic[word]+=1
            
    top=sorted(dic.items(), key=lambda x:x[1],reverse=True)[:10] 
    x,y=zip(*top)
    plt.bar(x,y)
```

```python
plot_top_stopwords_barchart(data['DESCRIPTION'])
```

### Note: Ngram/trigram exploration

```python
from nltk.util import ngrams
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def get_top_ngram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(n, n)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) 
                  for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:10]

top_n_bigrams=get_top_ngram(data['DESCRIPTION'],2)[:10]
x,y=map(list,zip(*top_n_bigrams))
sns.barplot(x=y,y=x)
```

```python
top_tri_grams=get_top_ngram(data['DESCRIPTION'],n=3)
x,y=map(list,zip(*top_tri_grams))
sns.barplot(x=y,y=x)
```

## Note: Stemming/lemmatization/tokenize words

```python
nltk.download('punkt')
nltk.download('wordnet')
```

```python
from nltk.stem.porter import *
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize
import gensim
```

```python
def preprocess_reports(df):
    corpus=[]
    stem=PorterStemmer()
    lem=WordNetLemmatizer()
    for reports in data['DESCRIPTION']:
        words=[w for w in word_tokenize(reports) if (w not in stop)]
        
        words=[lem.lemmatize(w) for w in words if len(w)>2]
        
        corpus.append(words)
    return corpus

corpus=preprocess_reports(reports)
```

```python
dic=gensim.corpora.Dictionary(corpus)
bow_corpus = [dic.doc2bow(doc) for doc in corpus]
```

```python
lda_model = gensim.models.LdaMulticore(bow_corpus, 
                                   num_topics = 4, 
                                   id2word = dic,                                    
                                   passes = 10,
                                   workers = 2)
lda_model.show_topics()
```

## Exploring entities with spacy 

```python
import spacy 
```

```python
from spacy import displacy
```

```python
import en_core_web_sm
```

```python
def plot_named_entity_barchart(text):
    nlp = spacy.load("en_core_web_sm")
    
    def _get_ner(text):
        doc=nlp(text)
        return [X.label_ for X in doc.ents]
    
    ent=text.apply(lambda x : _get_ner(x))
    ent=[x for sub in ent for x in sub]
    counter=Counter(ent)
    count=counter.most_common()
    
    x,y=map(list,zip(*count))
    sns.barplot(x=y,y=x)
```

```python
plot_named_entity_barchart(data['DESCRIPTION'])
```

```python
def plot_most_common_named_entity_barchart(text, entity="PERSON"):
    nlp = spacy.load("en_core_web_sm")
    
    def _get_ner(text,ent):
        doc=nlp(text)
        return [X.text for X in doc.ents if X.label_ == ent]

    entity_filtered=text.apply(lambda x: _get_ner(x,entity))
    entity_filtered=[i for x in entity_filtered for i in x]
    
    counter=Counter(entity_filtered)
    x,y=map(list,zip(*counter.most_common(10)))
    sns.barplot(y,x).set_title(entity)
```

```python
#plot persons - doesn't detect well at all.
plot_most_common_named_entity_barchart(data['DESCRIPTION'], entity="PERSON")
```
