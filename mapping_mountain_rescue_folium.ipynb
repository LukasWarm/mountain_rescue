## Clean lat/lng column for folium 


```python
df['Lat/Lng'] = df['Lat/Lng'].str.replace('POINT', '')
```

```python
df.rename(columns={"Lat/Lng": "LatLng"}, inplace=True)
```

```python
df.head()
```

```python
df[['Lng','Lat']] = df.LatLng.str.split(expand=True)
```

```python
df.head()
```

```python
df['Lat'] = df['Lat'].str.replace(')', '')
df['Lng'] = df['Lng'].str.replace('(', '')
```

```python
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
map_points = folium.Map(location=UK,
                      room_start=12,
                      tiles='CartoDB dark_matter')
for i in range (0,1128):
    folium.CircleMarker(location=[[df['Lat']][0][i], [df['Lng']][0][i]],
                       radius=2, stroke=False, fill_color='blue',
                                  fill_opacity=0.5).add_to(map_points)
```

```python
map_points
```

```python
map_cluster = folium.Map(location=UK,
                      room_start=12,
                      tiles='CartoDB dark_matter')
cluster = folium.plugins.MarkerCluster().add_to(map_cluster)
for i in range (0,1128):
    folium.Marker(location=[[df['Lat']][0][i], [df['Lng']][0][i]]).add_to(cluster)
map_cluster
```
