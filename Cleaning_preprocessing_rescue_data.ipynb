```python
injured_kw = '|'.join(['sustained','cut','tripped','slipped','dislocated','fracture', 'twisted', 'sprain', 'fell', 'fall', 'fallen', 'broken ankle', 'injured', 'injuries', 'injury', 'tumble'])
medical_issues_kw = '|'.join(['medical conditions','unwell','taken ill','became ill', 'collapsed'])
missing_kw = '|'.join(['vulnerable','last seen', 'missing', 'dazed'])
arm_kw = '|'.join(['arm'])
leg_kw = '|'.join(['leg','tibia', 'fibula', 'femur'])
hip_kw = '|'.join(['hip'])
elbow_kw = '|'.join(['elbow'])
soft_tissue_kw = '|'.join(['soft tissue'])
pelvis_kw = '|'.join(['pelvis'])
face_kw = '|'.join(['jaw', 'facial'])
wrist_kw = '|'.join(['wrist'])
head_kw = '|'.join(['head'])
neck_kw = '|'.join(['neck'])
back_kw = '|'.join(['lower back'])
hand_kw = '|'.join(['hand'])
shoulder_kw = '|'.join(['shoulder'])
knee_kw = '|'.join(['knee'])
cardiac_kw = '|'.join(['cardiac arrest'])
lost_kw = '|'.join(['overdue','disorientated', 'lost', 'separated', 'seperated'])
shouts_for_help_kw = '|'.join(['flares','flashing lights', 'called for help', 'false alarm', 'ran out of steam', 'unable', 'exhausted', 'nothing found', 'found nothing', 'false alarm'])
benighted_kw = '|'.join(['dark','torches', 'darkness', 'benighted'])
cragfast_kw = '|'.join(['cragfast'])





data['injured'] = data['DESCRIPTION'].str.contains(injured_kw).astype(int)
data['medical'] = data['DESCRIPTION'].str.contains(medical_issues_kw).astype(int)
data['missing'] = data['DESCRIPTION'].str.contains(missing_kw).astype(int)
data['arm'] = data['DESCRIPTION'].str.contains(arm_kw).astype(int)
data['leg'] = data['DESCRIPTION'].str.contains(leg_kw).astype(int)
data['hip'] = data['DESCRIPTION'].str.contains(hip_kw).astype(int)
data['elbow'] = data['DESCRIPTION'].str.contains(elbow_kw).astype(int)
data['soft tissue'] = data['DESCRIPTION'].str.contains(soft_tissue_kw).astype(int)
data['pelvis'] = data['DESCRIPTION'].str.contains(pelvis_kw).astype(int)
data['face'] = data['DESCRIPTION'].str.contains(face_kw).astype(int)
data['wrist'] = data['DESCRIPTION'].str.contains(wrist_kw).astype(int)
data['head'] = data['DESCRIPTION'].str.contains(head_kw).astype(int)
data['neck'] = data['DESCRIPTION'].str.contains(neck_kw).astype(int)
data['back'] = data['DESCRIPTION'].str.contains(back_kw).astype(int)
data['hand'] = data['DESCRIPTION'].str.contains(hand_kw).astype(int)
data['shoulder'] = data['DESCRIPTION'].str.contains(shoulder_kw).astype(int)
data['knee'] = data['DESCRIPTION'].str.contains(knee_kw).astype(int)
data['cardiac'] = data['DESCRIPTION'].str.contains(cardiac_kw).astype(int)
data['lost'] = data['DESCRIPTION'].str.contains(lost_kw).astype(int)
data['shouts_for_help'] = data['DESCRIPTION'].str.contains(shouts_for_help_kw).astype(int)
data['benighted'] = data['DESCRIPTION'].str.contains(benighted_kw).astype(int)
data['cragfast'] = data['DESCRIPTION'].str.contains(cragfast_kw).astype(int)
```
