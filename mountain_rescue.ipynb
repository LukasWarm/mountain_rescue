```python
#remove stop words
token_clean = [x for x in cat if x not in stopwords]
print(token_clean)
```

```python
#remove other tokens like ",", ";", ".", ...
punc = [",", ".", ";", " ", ":", "'", "..", "...", "&", "etc", "(", ")", "/", "_", "-", "’", "--", "``", "'s", "", "!"]
token_cleaner = [x for x in token_clean if x not in punc]
print(token_cleaner)
```

```python
fdist = FreqDist(token_cleaner)
fdist100 = fdist.most_common(100)
fdist100
```
